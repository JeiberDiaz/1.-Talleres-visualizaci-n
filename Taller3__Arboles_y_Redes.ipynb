## ***Abstracción***

### **¿Qué?**
Se observa un **dataset** de tipo **red**, el cual presenta diferentes relaciones existentes entre especies de la bahía de Florida; específicamente muestra la cadena alimenticia de las especies ubicadas en esta zona.

Se registran, por tanto, variables categoricas en parejas, siendo cada una de estas un nodo de la red. 

```python
print("Número de nodos:", g1.number_of_nodes())
print("Número de nodos:", g1.number_of_edges())
```

### **Visualización: Red de la relación entre especies de la bahía de Florida**

```python
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(g1, with_labels=True)
```

### **¿Por qué?**
> *Derivar / Explorar Topología*.

Determinar la forma en la cual se presentan las relaciones alimenticias de las especies existentes en la bahía de Florida. ¿Existe una forma determinada en estas relaciones? ¿Un patron? ¿Especies comunes o clusters en el comportamiento de dicha cadena? Cómo se relaciona cada especie, sus conexiones, entre otros.

### **¿Cómo?**
Encode (Separar) --> Navegar  --> Seleccionar --> Filtrar

En primer instancia se separa cada uno de los nodos, luego se navega con cada uno y se determinan los nodos con los que tiene relacion, los cuales se seleccionan para así generar la unión (juxtapocisión). Este proceso se realzia para cada nodo.

Dada la cantidad de nodos es muy recomendable (casi que **obligatorio**) generar animación para la realización de la red y evidenciar de mejor forma las diferentes relaciones generadas. 

## **Marcas y canales usados**
* Marca de punto para la representación de cada uno de los nodos existentes en la base.
* Marca de linea que indica la relación entre nodos; los enlaces entre nodos. 


## **Reglas de la visualización y aspectos a mejorar**
Se observa una evidente oclusión generada en el gráfico de la red dada la cantidad enorme de nodos y enlaces existentes. El aplicar un efecto 3D en este caso es justificado de haber aplicado una animación en la visualización. No son evidentes ni las relaciones entre nodos ni los nodos como tal; el texto que identifica a cada uno es ilegible para la gran mayoria de ellos. 

En general se cumple el cometido y se evidencian las relaciones buscadas, pero estos aspectos mencionados anteriorente serian determinantes para el correcto desarrollo de la actividad.
