## Visualización número 1.
El siguiente gráfico muestra el total de hectáreas utilizadas para el cultivo en el territorio nacional junto al rendimiento de dichos cultivos (total producción por hectárea) desde al año 2006 hasta el año 2018.

Recuperado de: https://www.datos.gov.co/Agricultura-y-Desarrollo-Rural/RENDIMIENTO-DE-CULTIVOS-EN-COLOMBIA-POR-A-O/fgh5-rjk

### **Abstracción**

#### *¿Qué?*
Los datos usados para la realización del gráfico provienen de la base histórica de la producción agrícola a nivel nacional, desagregando por departamentos, de los años 2007 a 2018. En ella se invcolucran diferentes varibles referentes a las características de la producción por departamento, pero en este caso se analizan específicamente tres variables: el año, el total del área sembrada a nivel nacional y el rendimiento de dichas areas sembradas en cuanto a producción se refiere.

Las caracteristicas del set de datos, el cual es representado en forma de tabla de datos *temporales* se da a continuación:

1. **Año**: Variable ordenada cualitativa, secuencial *(se supone)*, que representa la temporalidad de la medición. El rango de las mediciones se encuentra entre el 2006 y el 2018. 
2. **Área sembrada (ha)**: Variable ordenada cuantitativa que indica el total de hectáreas sembradas a nivel nacional en un año determinado. Su rango es comprendido por valores iguales o mayores a 0. (x >= 0)
3. **Rendimiento(t/ha)**: Variable cuantitativa ordenada que indica la producción total de los cultivos por hectárea. Su rango es comprendido por valores estrictamente mayores a 0. (x > 0)

#### *¿Por qué?*
**Explorar/Descubrir Tendencias**

**Comparar Similaridad**

A priori, se supone que se busca descubrir y comparar la tendencia que presenta el total de hectáreas sembradas frente al rendimiento general de los cultivos a nivel nacional en el periodo comprendido entre los años 2006 a 2018.

Dado que los datos se encuentran desagregados por departamento y tipo de producto se procede a agrupar el total de hectáreas sembradas y cosechadas por año. 

#### *¿Cómo?*
Se recurre a una codificación ordenada (por año, se supone) para poder analizar las tendencias del total de área de siembra y su respectivo rendimiento. Para separar estos dos items se emplea un color diferente. 
* Select → Juxtapose → filter.

### **Marcas y canales**

En general, las marcas y los canales usados en la representación visual son:
* Marca de área sobre el total de áreas sembradas en el terrotorio nacional.
* Marca de línea para el rendimiento total de los cultivos
* Canal de posición espacial vertical para representar el área sembrada
* Canal de posición espacial vertical para representar el rendimiento de los cultivos
* Canal de posición espacial horizontal para representar el año de medición
* Canal de color para separar los atributos: área sembrada y rendimiento de los cultivos
* Canales de pocisión de escala in-alineada. Esto dado que el área de siembra y rendimiento presentan escalas y rangos diferentes.

###**Inconvenientes de la visualización y aspectos a mejorar**

1. Las etiquetas de los años están "ordenados" de mayor a menor, empezando por el 2017 y terminando en el año 2006. Esto provoca inicalmente una confusion terrible al ver por primera vez el diagrama. Sin ver inicialmente estos niveles se supone una tendencia decreciente; implicaría que al avanzar el tiempo la cantidad de hectáreas sembradas y su rendimiento ha ido disminuyendo. Al observar los labels cambia totalmente el significado, presentando de hecho un aumento. Esto dado que el tiempo por lo general se ordena secuencialmente, de menor a mayor periodo. 

2. Existen dos años desordenados. Si bien es cierto que el orden de los años es descendente los años 2013 y 2018 se encuentran desordenados en el eje del tiempo. Sumando lo anterior la interpretabilidad se pierde en gran medida. No hay forma directa de determinar tendencias por ejemplo. Se pierde la precisión.

3. Sería buena idea incluir una leyenda o algún otro método para diferenciar las escalas de las dos variables cuantitativas presentes en el gráfico. Esto para evitar posibles confusiones al dar la misma escala a ambas variables. 

4. Situar mejor el recuadro de la leyenda de las variables. En esta ocasión el recuadro aparece a clickear una pestaña, pero de no ser el caso y saliese de esa forma por default sería mucho mejor ubicarlo en la parte superior derecha para evitar desperdiciar espacio.
